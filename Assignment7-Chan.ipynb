Normalizing Data
-----------------------------------------------
The CSV file was uploaded from the location of Downloads/movies.csv and read into the variable called "df"

The average rating for each movie was calculated using the mean() function and stored in "movie_avg"
The average rating for each user was calculated using the stack(), groupby(), and mean() function and stored in "user_avg"

A new DataFrame called "filledNans" was created to fill the empty Nan values with the value of Zero.
This process used the transform() and fillna() functions.

A new normalized DataFrame is created from filledNans using the normalization formula. 
The average rating for each movie and each user was again calculated and displayed.

The normalized ratings scaled all the numeric values into the range from 0 to 1.

This can be advantageous because many users may consistently give out high ratings, while other users consistently give out low ratings.  
By normalizing the ratings to scores between 0 and 1, the ratings outliers are removed and a more accurate rating is produced from a set of data where the user ratings are in a consistent scale from 0 to 1.
The few very high ratings or few very low ratings will no longer skew the average rating of the movie.

However, there are some disadvantages of using normalization. 
By shrinking the ratings between 0 and 1, the scale becomes much smaller and this can give out a limited opinion of the movies. 
Users will not be able to fully describe their opinions of good, average, or bad movies just based on a 0 to 1 rating. 
Users may feel that a 1 to 5 scale more accurately reflects the range of their opinions on a good, bad, or average movie.
The inclusion of decimal values may further confuse readers who want an easier to read 1 to 5 scale.


````python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
df = pd.read_csv('Downloads/movies.csv')

print (df)

movie_avg = df.mean()
print ("The Average Rating for each movie is: ")
print (movie_avg)

user_avg = df.set_index('User').stack().groupby('User').mean()
print ("The Average Rating given by each User is: ")
df = df.set_index('User')
user = df.mean(axis=1)
print (user_avg)

print ("Show the Normalized Ratings for each User: ")
grouped = df.groupby('User')

filled_Nans = grouped.transform(lambda x: x.fillna(df.mean(axis=1)))
print (filled_Nans)

norm_df = (filled_Nans - filled_Nans.min()) / (filled_Nans.max() - filled_Nans.min())
print (norm_df)

print ("The new average rating for each movie is now: ")
print (norm_df.mean())
print ("The new average rating for each user is now: ")
print (norm_df.mean(axis=1))

print ("Show the Standardized Ratings for each User: ")
std_df = ((filled_Nans - filled_Nans.mean()) / filled_Nans.std() )
print (std_df)

print ("The new average rating for each movie is now: ")
print (std_df.mean())
print ("The new average rating for each user is now: ")
print (std_df.mean(axis=1))
````
