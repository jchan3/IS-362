This python code reads the CSV file into a panda Data Frame variable called "df"

The Data Frame is into 2 different Data Frames called "alaska" for Alaska Airlines and "amwest" for AM West Airlines.

The code describes the On Time and Delayed statistics for Alaska and AM West.
The code then compares the average number delayed and on time arrivals for Alaska and AM West Airlines.
Lastly, the code comopares the sum of On Time and Delayed arrivals for Alaska and AM West Airlines.

Below is the Python code that can be run in the Jupyter Notebook:

````python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
df = pd.read_csv('Downloads/is362_project1.csv', header=None)

print (df)

df.columns = ['airline','city','on_time','delayed']

alaska = df[:5]
print (alaska)
amwest = df[5:10]
print (amwest)

alaska_avg_ontime = alaska['on_time'].mean()
alaska_avg_delayed = alaska['delayed'].mean()

amwest_avg_ontime = amwest['on_time'].mean()
amwest_avg_delayed = amwest['delayed'].mean()

alaska_describe_ontime = alaska['on_time'].describe()
print ("Describe On Time stats for Alaska Airlines: ")
print (alaska_describe_ontime)
amwest_describe_ontime = amwest['on_time'].describe()
print ("Describe On Time stats for AM West Airlines: ")
print (amwest_describe_ontime)

alaska_describe_delayed = alaska['delayed'].describe()
print ("Describe Delayed stats for Alaska Airlines: ")
print (alaska_describe_delayed)
amwest_describe_delayed = amwest['delayed'].describe()
print ("Describe Delayed stats for AM West Airlines: ")
print (amwest_describe_delayed)

alaska_sum_ontime = alaska['on_time'].sum()
alaska_sum_delayed = alaska['delayed'].sum()

amwest_sum_ontime = amwest['on_time'].sum()
amwest_sum_delayed = amwest['delayed'].sum()

print ("The average number of On Time Arrivals for Alaska Airlines is: ", alaska_avg_ontime)
print ("The average number of Delayed Arrivals for Alaska Airlines is: ", alaska_avg_delayed)

print ("The average number of On Time Arrivals for AM West Airlines is: ", amwest_avg_ontime)
print ("The average number of Delayed Arrivals for AM West Airlines is: ", amwest_avg_delayed)

print ("The total number of On Time Arrivals for Alaska Airlines is: ", alaska_sum_ontime)
print ("The total number of Delayed Arrivals for Alaska Airlines is: ", alaska_sum_delayed)

print ("The total number of On Time Arrivals for AM West Airlines is: ", amwest_sum_ontime)
print ("The total number of Delayed Arrivals for AM West Airlines is: ", amwest_sum_delayed)
   
````
