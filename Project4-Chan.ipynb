## Conclusion:
Based on the margin of error results, the "cap-shape" values of 0.4996104702538133, is a better predictor of the mushroom classification than the "odor" values of 0.500758529591673.

The lower the margin of error, the more accurate the attribute is at predicting the classification.

It is also possible to use both values or more than two values to create a better liinear regression model with better accuracy.  

````python
import numpy as np
import pandas as pd
from datetime import date
from pandas import DataFrame, Series
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import metrics
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LinearRegression

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data"
mushroom_df = pd.read_csv(url, header=None)
#print (mushroom_df.head())

#Create a new Data Frame copy
df = mushroom_df[[0,1,5]].copy()
print (df.head())

#Rename the dataframe columns
df.columns = ['class','cap-shape','odor']
print (df.head())

#Replace the codes in the data with numeric values
#change the class column
df["class"].replace(["p"], 1, inplace=True)
df["class"].replace(["e"], 0, inplace=True)
#change the cap-shape column
df["cap-shape"].replace(["b"], 0, inplace=True)
df["cap-shape"].replace(["c"], 1, inplace=True)
df["cap-shape"].replace(["f"], 2, inplace=True)
df["cap-shape"].replace(["k"], 3, inplace=True)
df["cap-shape"].replace(["s"], 4, inplace=True)
df["cap-shape"].replace(["x"], 5, inplace=True)
#change the odor column
df["odor"].replace(["a"], 0, inplace=True)
df["odor"].replace(["c"], 1, inplace=True)
df["odor"].replace(["f"], 2, inplace=True)
df["odor"].replace(["l"], 3, inplace=True)
df["odor"].replace(["m"], 4, inplace=True)
df["odor"].replace(["n"], 5, inplace=True)
df["odor"].replace(["p"], 6, inplace=True)
df["odor"].replace(["s"], 7, inplace=True)
df["odor"].replace(["y"], 8, inplace=True)
print (df.head())
print (df.dtypes)

#Create seaborn visualization
sns.pairplot(df, x_vars=['cap-shape','odor'], y_vars='class', kind='reg')

#Create new predictor columns for cap-shape
X = df[['cap-shape']]
y = df['class']
#Create test data with default setting of 25% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print (X_train.shape)
print (X_test.shape)
print (y_train.shape)
print (y_test.shape)
#Create linear regression model
linreg = LinearRegression()
#Fit linear regression model to training data
linreg.fit(X_train, y_train)
#Test the model to the test data
y_pred = linreg.predict(X_test)
#Calculate margin of error
capshape_results = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print (capshape_results)

#Create new predictor columns for odor
X = df[['odor']]
y = df['class']
#Create test data with default setting of 25% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print (X_train.shape)
print (X_test.shape)
print (y_train.shape)
print (y_test.shape)
#Create linear regression model
linreg = LinearRegression()
#Fit linear regression model to training data
linreg.fit(X_train, y_train)
#Test the model to the test data
y_pred = linreg.predict(X_test)
#Calculate margin of error
odor_results = np.sqrt(metrics.mean_squared_error(y_test, y_pred))
print (odor_results)
````
