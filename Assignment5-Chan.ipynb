The airports csv file and weather csv files were read from the URL web page into python data file variables.
The airport data file was sliced by Boolean selection to obtain the USA airport data.

The northern most airport was found by using the max function on the latitude column.
The eastern most airport was found by using the max function on the longitude column.

The weather data file was sliced by Boolean selection to find the weather on Feburary 12, 2013.
The windiest airport was found by using the max function on the wind speed column.

The top five most northern and eastern airports were found by sorting the data file by longitude and latitude with the sort_values function and displayed with the head() function. 


````python
import numpy as np
import pandas as pd
from datetime import date
from pandas import DataFrame, Series
temp = pd.read_csv('Downloads/airports.csv')

url = "https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/airports.csv"
df = pd.read_csv(url)

us = df[df.lon < 0]
max_lat = us['lat'].max()
northern_airport = df[df.lat == max_lat][['name']]

print ("The northern most airport in the USA is: ", northern_airport)
print ("The latitude value is: ", max_lat)

max_lon = us['lon'].max()
eastern_airport = df[df.lon == max_lon][['name']]

print ("The eastern most airport in the USA is: ", eastern_airport)
print ("The longitude value is: ", max_lon)

weather = "https://raw.githubusercontent.com/hadley/nycflights13/master/data-raw/weather.csv"
weather_df = pd.read_csv(weather)

m = 2
d = 12
y = 2013
weather_query = weather_df[(weather_df.month == m) & (weather_df.day == d) & (weather_df.year == y)]
max_wind = weather_query['wind_speed'].max()
windy_airport = weather_query[weather_query.wind_speed == max_wind][['origin']]
windy_faa = windy_airport.iat[0, 0]
airport_name = df[df.faa == windy_faa][['name']]
print ("The windiest New York airport on February 12, 2013 is: ", airport_name.iat[0, 0])
print ("The wind speed value was: ", max_wind)

print ("Here is a list of the top five most northern airports: ")
reindexed = us.sort_values('lat', ascending=False) 
five_north = reindexed.head()
print (five_north)

print ("Here is a list of the top five most eastern airports: ")
reindexed = us.sort_values('lon', ascending=False) 
five_east = reindexed.head()
print (five_east)

print ("Here is a list of the weather on February 12, 2013: ")
print (weather_query)
````
