The data file was uploaded from the url at "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data" and read into the variable called "df" with the delimeter of space separating the columns.

The DataFrame columns were renamed according to the attribute information provided in auto-mpg.names

The horsepower data with "?" was replaced with the NaN value using the replace function.
The horsepower column was converted to numeric using the apply function.

The origin column was converted to "USA", "Asia", and "Europe" using the replace function.

A bar chart was created showing the number of cars with cylinders using the plot function. 

A scatterplot was created showing the relationship between horsepower and weight using the plot function.

A multiple linear regression graph was created showing the weight as a function of mpg for each origin using the seaborn graphics capability lmplot function.

````python
import numpy as np
import pandas as pd
from datetime import date
from pandas import DataFrame, Series
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#load the data into a DataFrame
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data"
df = pd.read_csv(url, header=None, delimiter=r"\s+")
print (df.head())

#Rename the dataframe columns
df.columns = ['mpg','cylinders','displacement','horsepower','weight','acceleration','model_year','origin','car_name']
print (df.head())

#Replace ? with Nan and convert the column to Numeric
df["horsepower"].replace(["?"], np.NaN, inplace=True)
df["horsepower"] = df["horsepower"].apply(pd.to_numeric, errors='coerce')
print (df.dtypes)

#Convert the Origin Column values
df["origin"].replace([1], "USA", inplace=True)
df["origin"].replace([2], "Asia", inplace=True)
df["origin"].replace([3], "Europe", inplace=True)
print (df.head())

#Create a bar chart that shows the distribution for cylinders
cyl_df = df['cylinders'].value_counts()
cyl_df.plot(kind='bar');

#Create a scatterplot that shows the relationship between horsepower and weight
df.plot(kind='scatter', x='horsepower', y='weight');

#What is the distribution of the weight and mpg by country of origin?
sns.set()

# Plot weight as a function of mpg
g = sns.lmplot(x="mpg", y="weight", hue="origin", data=df)
````
