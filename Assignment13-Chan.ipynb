Preliminary Conclusions

The Data Frame contains data describing the class of mushroom as either poisonous or edible.
Based on the scatter plots, it appears that there is a relationship between poisonous mushrooms and the odor.

The poisonous mushrooms had strong odors of creosote, fishy, foul, musty, pungent, spicy, and none (1,2,4,5,6,7,8 values).

The edible mushrooms had pleasant odors of almond (a=0), anise (l=3), and none (n=5).


There is also a small relationship between the class of mushroom and the cap shape.

The edible mushrooms commonly had cap shape of bell, flat, sunken, and convex (0,2,4,5 values).

The poisonous mushrooms commonly had the cap shape of bell, conical, convex, flat, and knobbed (0,1,2,3,5 values).  
The poisonous mushrooms did not have "sunken" cap shapes.
The poisonous mushrooms also had more instances with the cap shapes of flat, knobbed, and convex (2,3,5 values).

````python
import numpy as np
import pandas as pd
from datetime import date
from pandas import DataFrame, Series
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/mushroom/agaricus-lepiota.data"
mushroom_df = pd.read_csv(url, header=None)
#print (mushroom_df.head())

#Create a new Data Frame
df = mushroom_df[[0,1,2,3,5]].copy()
print (df.head())

#Rename the dataframe columns
df.columns = ['class','cap-shape','cap-surface','cap-color','odor']
print (df.head())

#Replace the codes in the data with numeric values
#change the class column
df["class"].replace(["p"], 1, inplace=True)
df["class"].replace(["e"], 0, inplace=True)
#change the cap-shape column
df["cap-shape"].replace(["b"], 0, inplace=True)
df["cap-shape"].replace(["c"], 1, inplace=True)
df["cap-shape"].replace(["f"], 2, inplace=True)
df["cap-shape"].replace(["k"], 3, inplace=True)
df["cap-shape"].replace(["s"], 4, inplace=True)
df["cap-shape"].replace(["x"], 5, inplace=True)
#change the cap-surface column
df["cap-surface"].replace(["f"], 0, inplace=True)
df["cap-surface"].replace(["g"], 1, inplace=True)
df["cap-surface"].replace(["s"], 2, inplace=True)
df["cap-surface"].replace(["y"], 3, inplace=True)
#change the cap-color column
df["cap-color"].replace(["b"], 0, inplace=True)
df["cap-color"].replace(["c"], 1, inplace=True)
df["cap-color"].replace(["e"], 2, inplace=True)
df["cap-color"].replace(["g"], 3, inplace=True)
df["cap-color"].replace(["n"], 4, inplace=True)
df["cap-color"].replace(["p"], 5, inplace=True)
df["cap-color"].replace(["r"], 6, inplace=True)
df["cap-color"].replace(["u"], 7, inplace=True)
df["cap-color"].replace(["w"], 8, inplace=True)
df["cap-color"].replace(["y"], 9, inplace=True)
#change the odor column
df["odor"].replace(["a"], 0, inplace=True)
df["odor"].replace(["c"], 1, inplace=True)
df["odor"].replace(["f"], 2, inplace=True)
df["odor"].replace(["l"], 3, inplace=True)
df["odor"].replace(["m"], 4, inplace=True)
df["odor"].replace(["n"], 5, inplace=True)
df["odor"].replace(["p"], 6, inplace=True)
df["odor"].replace(["s"], 7, inplace=True)
df["odor"].replace(["y"], 8, inplace=True)
print (df.head())
print (df.dtypes)

#Perform exploratory data analysis
#Create a bar chart that shows the distribution for mushroom data
class_df = df['class'].value_counts()
shape_df = df['cap-shape'].value_counts()
surface_df = df['cap-surface'].value_counts()
color_df = df['cap-color'].value_counts()
odor_df = df['odor'].value_counts()

#Create Distribution Data Frame
dist_df = pd.concat([class_df, shape_df, surface_df, color_df, odor_df], axis=1)
dist_df.columns = ['class','cap-shape','cap-surface','cap-color','odor']
print (dist_df)
dist_df.plot(kind='bar');

#Create a scatterplot that shows the relationship between class and odor
scatter1 = df.plot(kind='scatter', x='class', y='odor')
print(scatter1)
scatter2 = df.plot(kind='scatter', x='class', y='cap-shape')
print(scatter2)
#ax1 = df.plot(kind='scatter', x='class', y='odor', color='r')    
#ax2 = df.plot(kind='scatter', x='class', y='cap-shape', color='g', ax=ax1)    
#ax3 = df.plot(kind='scatter', x='class', y='cap-surface', color='b', ax=ax1)
#ax4 = df.plot(kind='scatter', x='class', y='cap-color', color='y', ax=ax1)
#print(ax1 == ax2)

#What is the distribution of the odor and shape by class?
sns.set()

# Plot weight as a function of mpg
g = sns.lmplot(x="odor", y="cap-shape", hue="class", data=df)
````
